# Nested Modal .
---
Different types of information is stored in **Address** . For example , we need street number so we perform different steps to extract this information . <br>
So nested modal is used to solve this type of problems 

```python
#Create a pydantic modal for Address then we use as field in User modal. 
from pydantic import BaseModel
class Address(BaseModel):
    street: str
    city: str
    state: str
    zip_code: str
```

```python
from pydantic import BaseModel

class User(BaseModel):
    name : str
    age : int
    address : Address  # Using Address as a field in User modal
```

```python
#Example of creating a User instance with nested Address 

user_address = { 
    "street": "123 Main St",
    "city": "Anytown",
    "state": "CA",
    "zip_code": "12345"
}


user = Address(**user_address)  # Creating an Address instance
```

```python
# Now we can create a User instance using the Address instance
user_info = {
    "name": "John Doe",
    "age": 30,
    "address": user  # Using the Address instance in User modal
}

user_1 = User(**user_info)  # Creating a User instance
print(user_1)
```

```python
print(user_1.address.street)  # Accessing nested Address field in User modal
```
