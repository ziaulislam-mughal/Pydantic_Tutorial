# Computed Field 

Computed Field is calculated from two fields . Its is not input directly by the user . 

```python
from  pydantic import BaseModel , computed_field
from typing  import List , Dict , Optional 


class Patient(BaseModel):
    name: str
    age: int
    weight : float
    height : float
    married : bool 
    allergies : List[str]
    contact : Dict[str,str]

    #computed field  
    # Lets calculated Bmi on basis of Weight and height 

    @computed_field
    @property 
    def bmi(self)->float :
        bmi = round(self.weight / (self.height/100)**2,2)
        return f"The bmi of {self.name} is {bmi}"

    
```

```python

patient_info = {"name":"John Doe", 
                "age":65 , 
                "weight": 70.5,
                "height" : 175.0,
                "married": True ,
                "allergies": ["Peanuts", "Shellfish"],
                "contact": {"Phone": "03495623102", "emergence": "03432342334" , "Email":"ziamughal132@gmail.comp"}
                }


patient1 = Patient(**patient_info)

print(patient1)
```
