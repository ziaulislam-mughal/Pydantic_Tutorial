## The Pydantic Workflow

---

**Step 1. Define the Pydantic Model**

First, you build a Pydantic class to define your ideal data structure. You specify the required fields, such as name and age, clearly declare their expected data types (like string or integer), and apply specific validation rules, such as ensuring the age is a positive number.


### Example


```python
#Simple function of inserted
def inserted(name,age):
    print(name)
    print(age)
```

# Type Validation


```python
from pydantic import BaseModel
#Typing module for type hinting , List and Dict are string and also items in list and dict are also string. 
from typing import List , Dict

class Patient(BaseModel):
    name: str
    age: int
    weight : float
    married : bool 
    allergies : List[str]
    contact : Dict[str,str]

def inserted(name,age):
    print(name)
    print(age)


```

**Step 2: Create and Validate the Object**

Next, you take raw input data—for example, a dictionary like {"name": "ali", "age": 30}—and use it to instantiate your Pydantic class. Pydantic automatically checks this raw input against your validation rules (verifying the age is positive) and successfully converts it into a clean, validated Pydantic object.


```python
patient_info = {"name":"John Doe", 
                "age":30 , 
                "weight": 70.5,
                "married": True ,
                "allergies": ["Peanuts", "Shellfish"],
                "contact": {"Phone": "03495623102", "Email":"ziamughal132@gmail.comp"}
                }
patient1 = Patient(**patient_info)
```

```python
patient_info_2 = {"name"    :   "Ali khan", 
                "age"       :   43 , 
                "weight"    :   33.5,
                "married"   :   False ,
                "allergies" :   ["corona", "flu"],
                "contact"   :   {"Phone": "0349897297", "Email":"noalikhanhere232@gmail.comp"}
                }
```

```python
patient2 = Patient(**patient_info_2)
```

**Step 3: Process the Validated Data**

Finally, you pass this fully validated Pydantic object into a function. Because the data is already validated, the function can safely and reliably implement its entire business logic without worrying about missing or incorrect data.


```python
def inserted(patient: Patient):
    print(patient.name)
    print(patient.age)
    print(patient.weight)
    print(patient.married)
    print(patient.allergies)
    print(patient.contact)
    print("inserted successfully")
```

```python
inserted(patient1)
```

```python
inserted(patient2)
```

### Required Fields & Optional Fields. <br>
By default fields are **required** . Suppose we make optional allergies field. 

```python
from pydantic import BaseModel
from typing import List , Dict , Optional

class Patient(BaseModel):
    name: str
    age: int
    weight : float
    married : bool 
    #Allergies are optional column and give default value of None if not provided.
    allergies : Optional[List[str]] = None 
    contact : Dict[str,str]
```

```python
patient_info_2 = {"name"    :   "Ali khan", 
                "age"       :   43 , 
                "weight"    :   33.5,
                "married"   :   False ,
                "contact"   :   {"Phone": "0349897297", "Email":"noalikhanhere232@gmail.comp"}
                }


patient2 = Patient(**patient_info_2)
```

```python
inserted(patient2) # Ouptut will be None for allergies because it is optional and not provided in patient_info_2
```
