## The Pydantic Workflow 
---

**Step 1.  Define the Pydantic Model**

First, you build a Pydantic class to define your ideal data structure. You specify the required fields, such as name and age, clearly declare their expected data types (like string or integer), and apply specific validation rules, such as ensuring the age is a positive number.



### Example 

```python
#Simple function of inserted
def inserted(name,age):
    print(name)
    print(age)
```

### Type Validation 

```python
from pydantic import BaseModel
class Patient(BaseModel):
    name: str
    age: int


def inserted(name,age):
    print(name)
    print(age)


```

**Step 2: Create and Validate the Object**

Next, you take raw input data—for example, a dictionary like {"name": "ali", "age": 30}—and use it to instantiate your Pydantic class. Pydantic automatically checks this raw input against your validation rules (verifying the age is positive) and successfully converts it into a clean, validated Pydantic object.



```python
patient_info = {"name":"John Doe", "age":30}
patient1 = Patient(**patient_info)
```

**Step 3: Process the Validated Data**

Finally, you pass this fully validated Pydantic object into a function. Because the data is already validated, the function can safely and reliably implement its entire business logic without worrying about missing or incorrect data.

```python
def inserted(patient: Patient):
    print(patient.name)
    print(patient.age)
    print("inserted successfully")
```

```python
inserted(patient1)
```
