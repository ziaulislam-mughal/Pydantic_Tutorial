### Type Validation 
we write a insert funstion , which takes 2 arguments . <br>
- name
- age <br>

In python , datatype is dynamically select . When data is input in variable , according to data , datatype is selected . 
**Problem** <br>
What if user input **thirty** instead of 30 . 

```python
# insert function 
def insert(name,age):
    print("Name:", name)
    print("Age:", age)
```

```python
insert("Alice", 'thirty')
```

**Solution** <br>
Manually , write code , add conditions to control type validation .eg **Type hunting**

```python
# write instruction ,while input data , Enter the name  should string and age should be integar.

def insert(name, age):
    if type(name) == str and type(age) == int:
        print("Name:", name)
        print("Age:", age)
```

# Data Validation 
if user input negative age . 

```python
insert("Bob", -25)
```

**Solution** <br>
Add condition to control it .

```python
def insert(name, age):
    if type(name) == str and type(age) == int and age >= 0:
        print("Name:", name)
        print("Age:", age)
```
