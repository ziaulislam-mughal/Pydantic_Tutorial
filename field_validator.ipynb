# 1. Custom Data Validation 

### Situation 
Hospital has connection with banks . Employess of Bank are eligible to get medicie or service of hospital in reduce amount . <br>
Now check Patient is employee of Bank are or  not . 
#### How we check . 
Employess has special email . check **hdfc.com** if their is gmail.com . he is not valid . 
This is also data validation . It is specific bussiness use case . 

Import Field Validator from pydantic . 

```python
from pydantic import field_validator
```

```python
from pydantic import BaseModel , Field , EmailStr , AnyUrl 
from typing import Optional , List , Dict , Annotated
```

Define a method which check emails . Method use with decorater . called field_validor .

```python
class Patient(BaseModel):
    name: str
    age: int
    email : EmailStr
    LinkedIn : Optional[AnyUrl] = None
    weight : float
    married : bool 
    allergies : List[str]
    contact : Dict[str,str]



    #Apply which field 
    @field_validator('email')
    #Tell it is class method
    @classmethod
    #now define method 
    def email_validator(cls, value):
        valid_domain = "hdfc.com"
        #abc@gmail.com
        domain = value.split("@")[-1]
        if domain != valid_domain:
            raise ValueError(f"Email domain must be {valid_domain}")

        return value

```

```python
user_info = {
    "name": "Ali" ,
    "age": 20 ,
    "email" : 'alikhan@hdfc.com' , #"alikhan@gmail.com - this will raise error"
    "weight" : 34.3,
    "married" : True ,
    "allergies" : ['dust' , 'pollen' , 'nuts'] ,
    "contact" : {"Phone": "1234567890" , "Address" : "123 Main St"}
}
```

```python
user = Patient(**user_info)
```

```python
print(user)
```

# 2. Transformation

### Situation 
name field must be upper case . 

```python
class Patient(BaseModel):
    name: str
    age: int
    email : EmailStr
    LinkedIn : Optional[AnyUrl] = None
    weight : float
    married : bool 
    allergies : List[str]
    contact : Dict[str,str]


    @field_validator('name')
    @classmethod
    def transform_name(cls, value):
        return value.upper()
        
```

```python
user_info = {
    "name": "Ali" ,
    "age": 20 ,
    "email" : 'alikhan@hdfc.com' , #"alikhan@gmail.com - this will raise error"
    "weight" : 34.3,
    "married" : True ,
    "allergies" : ['dust' , 'pollen' , 'nuts'] ,
    "contact" : {"Phone": "1234567890" , "Address" : "123 Main St"}
}
```

```python
user2 = Patient(**user_info)
print(user2)
```
