```python
from pydantic import BaseModel
from typing import List , Dict , Optional

class Patient(BaseModel):
    name: str
    age: int
    weight : float
    married : bool 
    #Allergies are optional column and give default value of None if not provided.
    allergies : Optional[List[str]] = None 
    contact : Dict[str,str]

patient_info_2 = {"name"    :   "Ali khan", 
                "age"       :   43 , 
                "weight"    :   33.5,
                "married"   :   False ,
                "contact"   :   {"Phone": "0349897297", "Email":"noalikhanhere232@gmail.comp"}
                }


patient2 = Patient(**patient_info_2)

def inserted(patient: Patient):
    print(patient.name)
    print(patient.age)
    print(patient.weight)
    print(patient.married)
    print(patient.allergies)
    print(patient.contact)
    print("inserted successfully")

inserted(patient2) # Ouptut will be None for allergies because it is optional and not provided in patient_info_2
```
