## Email Validator . 

```python
#!pip install email-validator
```

```python
from pydantic import BaseModel , EmailStr 


class User(BaseModel):
    name : str 
    email : EmailStr 



user_info = {"name" : "Ali kiani", "email" : "alikhan@gmail.com"}

user = User(**user_info)

print(user)
```

## URL Validator

```python
from pydantic import BaseModel , EmailStr ,AnyUrl


class User(BaseModel):
    name : str 
    email : EmailStr 
    #linkedin id URL 
    linkedin : AnyUrl



user_info = {"name" : "Ali kiani", "email" : "alikhan@gmail.com" , "linkedin" : "https://www.linkedin.com/in/alikiani/"}

user = User(**user_info)

print(user)
```

## Custom Data Validation 

**Field Function** <br>
Suppose , user does not input negative value of age field. 

```python
from pydantic import Field 

class User(BaseModel):
    #age must greater than zero 
    age : int = Field(..., 
                      gt=0 ,
                      description="Age must be greater than 0")
    #name must be at least 5 characters long
    name : str = Field(..., 
                       min_length=5 , 
                       description="Name must be at least 5 characters long")
    
user_info = {"age" : 25, "name" : "Ali kiani"}
user = User(**user_info)
print(user)
```

**Field Function** is also used to attach **Meta data** . <br>
We use combination of **Field** and **Annotated** to attach metadata . <br>
We can also set default value . <br>
**Strict** is used , when we not allowed conversion. <br>
Pydantic is smart enough to understand "75.5" is float . <br>
But it is written in commas . which represent it string . so we use strict . so user add number , float . 

```python
from typing import Annotated 
from pydantic import Field 

class User(BaseModel):
    #Name
    name : Annotated[str , 
                    Field(max_length=20 ,
                    description="Name must be less than 20 characters long",
                    title="User Name",
                    examples=["Ali kiani" , "John Doe"])]
    #married 
    married : Annotated[bool , Field(
        default=None , 
        description="Marital status of the user",
        title="Married",
    )]

    #weight 
    weight : Annotated[float , Field(
        strict=True , #strict use to ensure that the value is of the correct type and not coerced from another type.
        gt=0 ,
    )]
   
    
user_info = {"name" : "Ali kiani" , "weight" : 70.5}
user = User(**user_info)
print(user)
```
